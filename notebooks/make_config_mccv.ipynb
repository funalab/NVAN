```python
import os
import sys
from glob import glob
```

```python
model = ['LSTM', 'LSTMAttention', 'LSTMMultiAttention', 'MuVAN', 'NVAN']

# {kfcv, mccv}
cv = 'mccv'

cv_list = glob('/Users/tokkuman/git-tokkuman/embryo_classification/datasets/split_list/{}/set*'.format(cv))
root = '/Users/tokkuman/git-tokkuman/embryo_classification/confs/models/'
os.makedirs(os.path.join(root, cv), exist_ok=True)

for m in model:
    os.makedirs(os.path.join(root, cv, m), exist_ok=True)
    for c in cv_list:
        dist = os.path.join(root, cv, m, 'train_{}.cfg'.format(os.path.basename(c)))
        with open(dist, 'w') as f:
            f.write('[Dataset]\n')
            f.write('root_path = datasets\n')
            f.write('split_list_train = datasets/split_list/{0}/{1}/train.txt\n'.format(cv, os.path.basename(c)))
            f.write('split_list_validation = datasets/split_list/{0}/{1}/validation.txt\n'.format(cv, os.path.basename(c)))
            f.write('basename =input\n')
            f.write('\n')

            f.write('[Model]\n')
            f.write('model = {}\n'.format(m))
            f.write('# init_classifier =\n')
            f.write('input_dim = 11\n')
            f.write('num_classes = 2\n')
            f.write('num_layers = 2\n')
            f.write('hidden_dim = 128\n')
            f.write('dropout = 0.5\n')
            if m in [None]:
                f.write('lossfun = nn.CrossEntropyLoss()\n')
            elif m in ['LSTM', 'LSTMAttention', 'LSTMMultiAttention', 'MuVAN', 'NVAN']:
                f.write('lossfun = nn.BCEWithLogitsLoss()\n')
            else:
                sys.exit()
            f.write('eval_metrics = f1\n')
            f.write('\n')

            f.write('[Runtime]\n')
            f.write('save_dir = results/train_{0}_set{1:02d}\n'.format(m, int(os.path.basename(c)[3:])))
            f.write('batchsize = 2\n')
            f.write('val_batchsize = 1\n')
            f.write('epoch = 100\n')
            f.write('optimizer = Adadelta\n')
            f.write('lr = 1.0\n')
            f.write('momentum = 0.95\n')
            f.write('weight_decay = 0.001\n')
            f.write('delete_tp = 50\n')
            f.write('# cuda:0 or cpu\n')
            f.write('device = cuda:0\n')
            f.write('seed = 0\n')
            f.write('phase = train\n')
            f.write('graph = False\n')

        dist = os.path.join(root, cv, m, 'test_{}.cfg'.format(os.path.basename(c)))
        with open(dist, 'w') as f:
            f.write('[Dataset]\n')
            f.write('root_path = datasets\n')
            f.write('split_list_test = datasets/split_list/{0}/{1}/test.txt\n'.format(cv, os.path.basename(c)))
            f.write('basename =input\n')
            f.write('\n')

            f.write('[Model]\n')
            f.write('model = {}\n'.format(m))
            f.write('init_classifier =\n')
            f.write('input_dim = 11\n')
            f.write('num_classes = 2\n')
            f.write('num_layers = 2\n')
            f.write('hidden_dim = 128\n')
            f.write('dropout = 0.5\n')
            if m in [None]:
                f.write('lossfun = nn.CrossEntropyLoss()\n')
            elif m in ['LSTM', 'LSTMAttention', 'LSTMMultiAttention', 'MuVAN', 'NVAN']:
                f.write('lossfun = nn.BCEWithLogitsLoss()\n')
            else:
                sys.exit()
            f.write('eval_metrics = f1\n')
            f.write('\n')
            f.write('[Runtime]\n')
            f.write('save_dir = results/test_{0}_set{1:02d}\n'.format(m, int(os.path.basename(c)[3:])))
            f.write('val_batchsize = 1\n')
            f.write('# cuda:0 or cpu\n')
            f.write('device = cuda:0\n')
            f.write('seed = 0\n')
            f.write('phase = test\n')
```
